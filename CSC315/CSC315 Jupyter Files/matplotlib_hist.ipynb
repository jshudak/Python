```python
#%matplotlib notebook
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
```

```python
A = np.random.randint(0,4, size=10)  # size=10 is a keyword argument to the function randint
A
```

```python
fig1, ax1 = plt.subplots(1,1)
ax1.hist(A)
ax1.set_title('histogram of A')
```

```python
A_rand = np.random.rand(1000)
fig2, ax2 = plt.subplots(1,1)
ax2.hist(A_rand)
ax2.set_title('histogram of A_rand - Uniform')
```

```python
A_randn = np.random.randn(1000)
fig3, ax3 = plt.subplots(1,1)
ax3.hist(A_randn)
ax3.set_title('histogram of A_randn - Gaussian')
```

```python
fig4, (ax1, ax2) = plt.subplots(1,2)
ax1.hist(A_rand)
ax1.set_title('A_rand - Uniform')
ax2 = fig4.add_subplot(122)
ax2.hist(A_randn)
ax2.set_title('A_randn - Gaussian')
```
