We can use 2D numpy array to store the point cooridates of object geometires, e.g., a triangle

```python
Triangle=np.array([[-0.5, 0],
                   [ 0.5, 0],
                   [ 0,   1]], dtype='float32')
```

Triangle is a 3 by 2 matrix, each row contains the x-y cooridate of apoint

using index notation ArrayName[:, k] to get a col with index k

```python
x = Triangle[:,0] # the first col
y = Triangle[:,1] # the second col
```

We can use a 2D numpy array to store the point cooridates of a Spiral

```python
Spiral=np.zeros((1000,2), dtype='float32') # 1000 points
Spiral.shape
```

```python
for n in range(0, 1000):
    theta = np.pi*2*n/100
    x = n*np.cos(theta)
    y = n*np.sin(theta)
    Spiral[n,0]=x
    Spiral[n,1]=y        
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(Spiral[:,0], Spiral[:,1], '-r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
```

We can use 2D numpy array to store the pixels of a 2D image <br>
Let's create an image of a tic-tac-toe gameboard

```python
I = np.array([[1,0,1],
              [0,1,0],
              [1,0,1]], dtype='float32')
```

```python
fig, ax = plt.subplots(1, 1)
ax.imshow(I, cmap='gray', vmin=0, vmax=1)
ax.set_aspect('equal')
```

Let's create an image of noises. We may see this image when a TV has no signal ...

```python
NoiseImage = np.random.rand(128, 256) # a 128 by 256 matrix, 2D array
fig, ax = plt.subplots(1, 1)
ax.imshow(NoiseImage, cmap='gray', vmin=0, vmax=1, origin='lower')
ax.set_axis_off() # hide the axes
ax.set_title('Your TV has no signal, check the cable ...')
```

```python
fig, ax = plt.subplots(1,1)
ax.hist(NoiseImage.reshape(-1))
ax.set_title('histogram of the image')
```
