# Plot Two-Dimensional (2D) lines/curves using Matplotlib

We can use Matplotlib to visualize a function with one variable, e,g. y = f(x)

````python
%matplotlib inline
#%matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
````

First, we create a figure (fig) with axes (ax) using ``plt.subplots``

````python
fig, ax = plt.subplots()
````

The *figure* (fig from plt.subplots()) is a container that holds the objects representing axes, graphics, text, and labels. <br>
The *axes* (ax from plt.subplots()) gives us a coordinate system: a bounding box with horizontal and vertical lines <br>
Once we have created a *fig* with *ax*, we can use the ``ax.plot`` function to plot some data

````python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
````

````python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel('x [time]', fontsize=20) # change x to x [time], fontsize to 30
ax.set_ylabel('y [amp]', fontsize=20) # change y to 'y [amplitude]'
ax.set_title('signal y = sin(x)', fontsize=20) # change y to 'y [amplitude]'
````

If we want to create a single figure with multiple lines, we can simply call the ``plot`` function multiple times:

We can plot multiple curves together in one axes/region

````python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, x, y2)
ax.set_xlabel('x', fontsize=20) # change fontsize from 20 to 100
ax.set_ylabel('y1 and y2', fontsize=20)
ax.set_title('y1=sin(x), y2=cos(x)', fontsize=20)
````

## Set Line Colors in a Plot

In the above example, we plot a sin curve, and a cos curve <br>
Each curve is assigned with a random color <br>
We can specify the color of each curve <br>
We can also change  the style of a curve (which is called line in a plot)

````python
#create signals
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#create a figure with axes
fig, ax = plt.subplots()
#set xlabel and ylabel
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
#
ax.plot(x, y0, color='blue')
#ax.plot(x, y0, 'b')
ax.plot(x, y1, color='green')
#ax.plot(x, y1, 'g')
ax.plot(x, y2, color='red')
#ax.plot(x, y2, 'r')
ax.plot(x, y3, color=(1, 1, 0)) # RGB color (red, green blue), the value is from 0 to 1
````

If no color is specified for a curve, Matplotlib will assign a color randomly

#### use one line of code to plot the four curves

````python
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#
fig, ax = plt.subplots()
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2', fontsize=20)
# one line of code to plot the four curves
ax.plot(x, y0, 'b', x, y1, 'g', x, y2, 'r', x, y3, 'y')
````

## Set Line Styles in a Plot

````python
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#
fig, ax = plt.subplots()
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
#
ax.plot(x, y0, color='blue', linestyle='solid')
#ax.plot(x, y0, 'b-')
ax.plot(x, y1, color='green', linestyle='dashed')
#ax.plot(x, y1, 'g--')
ax.plot(x, y2, color='red', linestyle='dashdot')
#ax.plot(x, y2, 'r-.')
ax.plot(x, y3, color='r', linestyle='dotted')
#ax.plot(x, y3, 'y:')
````

#### use one line of code to plot four curves with different colors and styles

````python
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#
fig, ax = plt.subplots()
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
# plot four curves in one line of code using different colors and styles
ax.plot(x, y0, 'b-', x, y1, 'g--', x, y2, 'r-.', x, y3, 'y:')
````

## Set Axes Limits of a Plot

We can set the ranges of the horizontal (x) axis and the vertical (y) axis <br>
using the functions ``ax.set_xlim()`` and ``ax.set_ylim() or ax.axis()``

````python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
#ax.set_xlim(-1, 10)
#ax.set_ylim(-10, 10)
#one line of code to set the two ranges
ax.axis([-1, 10, -2, 2])
````

We can adjust the aspect ratio between the two axes using the function ``ax.set_aspect``

````python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.axis([-1, 10, -10, 10])
ax.set_aspect(0.1) # change to other numbers 
#ax.set_aspect('equal')
````

## Add a legend (caption) in a Plot

We can add a legend in a plot to show the name of a curve

````python
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
fig, ax = plt.subplots()
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
ax.plot(x, y0, color='blue',   linestyle='solid',   label = 'y0=sin(x)') # set the label
ax.plot(x, y1, color='green',  linestyle='dashed',  label = 'y1=sin(x+1)')
ax.plot(x, y2, color='red',    linestyle='dashdot', label = 'y2=sin(x+2)')
ax.plot(x, y3, color='yellow', linestyle='dotted',  label = 'y3=sin(x+3)')
ax.axis([-1, 20, -1, 1])
#show the legend
ax.legend(loc='upper right', fontsize=16)
#add a tile
ax.set_title('Four curves in one plot', fontsize=16)
````

# Three-Dimensional (3D) Plot in Matplotlib
Matplotlib was initially designed for 2D visualization. <br>
After the release of v1.0, Matplotlib has a limited number of functions for 3D visualization in the module mplot3d

````python
from mpl_toolkits import mplot3d
````

````python
fig = plt.figure()
ax = plt.axes(projection='3d')
# the argument "projection='3d'" is needed to create a 3D axes
````

## Plot 3D Points and Lines

````python
#point data of a 3D curve
z = np.linspace(0, 10, 100)
x = np.sin(z)
y = np.cos(z)
#plot the 3D curve
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x, y, z, 'r-') # change '.' to '-' to show the curve
ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)
ax.set_zlabel('Z', fontsize=16)
ax.set_title('a trajectory of an object in a 3D space', fontsize=16)
````

## Plot 3D  Surfaces

````python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=False)
````

### Exercise 1

Implement a function according to the specification:
```python
def draw(t, x1, x2, c, fs):
    # parameters:
    #   t is a 1D array of timepoints (e.g. np.linspace(0, 10, 100))
    #   x1 is a 1D array of signal amplitude values (e.g. x1 is sin(t))
    #   x2 is a 1D array of signal amplitude values (e.g. x2 is cos(t))
    #   c is the color for plot
    #   fs is the fontsize of x_label, y_label, and title
    # 
    # specification:
    #   compute the average of the two signals: x = (x1+x2)/2
    #   plot the signal x
    #   set x_label to 't [time]'
    #   set y_label to 'x [signal amplitude]'
    #   set title to 'x(t) = (x1(t) + x2(t))/2'
    #   set color to c, set fontsize to fs
```
A test case: <br>
t = np.linspace(0, 10, 100)  <br>
x1 = np.sin(t)  <br>
x2 = np.cos(t)  <br>
draw(t, x1, x2, c='r', fs=16)

````python
#implement the function
def draw(t, x1, x2, c, fs):
    x = (x1+x2)/2
    fig,ax=plt.subplots()
    ax.plot(t,x,c)
    ax.set_xlabel('t [time]', fontsize=fs)
    ax.set_ylabel('x [signal amplitdue]', fontsize=fs)
    ax.set_title('x(t)=(x1(t)+x2(t))/2', fontsize=fs)
````

````python
#test the function
t = np.linspace(0, 10, 100) 
x1 = np.sin(t) 
x2 = np.cos(t) 
draw(t, x1, x2, c='r', fs=16)
````
